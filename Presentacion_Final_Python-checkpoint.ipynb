* Es importante obtener una copia de los encabezados predeterminados que usaran las solicitudes de descarga web para evitar el bloqueo. Adjunto está la solicitud de descarga web.

```python
headers = requests.utils.default_headers()
headers.update(
    {
        'User-Agent': 'Mozilla/5.0 (Linux; <Android Version>; <Build Tag etc.>) AppleWebKit/<WebKit Rev>(KHTML, like Gecko) Chrome/<Chrome Rev> Safari/<WebKit Rev>',
    }
)

html = requests.get('https://es.wikipedia.org/wiki/Primera_Divisi%C3%B3n_de_Espa%C3%B1a_2018-19',
                 headers = headers)
page = html.content
```

* Analizando el HTML descargado, creando estructura de arbol se extrae las tablas dentro, luego las filas y su contenido, para luego _`seleccionar solo los datos deseados`_.(_PUNTO DE CONTROL:_ Nos aseguramoes que en cada ejecución de este bloque de código **"page"** contenga los datos html iniciales con la primera linea)

```python
page = html.content  ## PUNTO DE CONTROL : Nos aseguramoes que en cada ejecución de este bloque de código "PAGE" contenga los datos html iniciales
page = BeautifulSoup(page, 'html.parser')
tablas = page.find_all('table', class_='sortable')
filas = [x for tabla in tablas for x in tabla.find_all('tr')]
contenido = [[x.text for x  in fila.find_all('td')] for fila in filas]
datos1 = pd.DataFrame(contenido).iloc[18:38,0:6] 
#datos ## primera vista de datos
```

* Empezamos con la limpieza y transformación de datos renombrando las columnas (_PUNTO DE CONTROL:_ Reasignando los datos iniciales de 1ra ejecución con la primera linea)

```python
datos = pd.DataFrame(contenido).iloc[18:38,0:6]  ## Control de ejecución: reasignando los datos iniciales de 1ra ejecución
datos.rename(columns={0:'NOM_EQUIPO',
                      1:'CIUDAD',
                      2:'ENTRENADOR',
                      3:'CAPITAN',
                      4:'ESTADIO',
                      5:'AFORO'
                      }, 
             inplace=True)
datos1 = datos.copy()   ## generamos una copia de los datos originales antes de la limpieza
datos.head()
```

* A continuación procedemos a limpiar y dar formato a todas las columnas del `dataframe` (_PUNTO DE CONTROL:_ reasignamos los valores iniciales al dataframe DATOS desde la copia de respaldo que se generó en el paso anterior)

```python
## Quitar restos de los datos en las columnas que no corresponden a los valores originales ejm. /n, /t, etc.
##============================================================================

datos = datos1.copy()   ## PUNTO DE CONTROL PARA EVITAR ERRORES Y PODER EJECUNAR N VECES

#---------------------------------------------------------------------------------------

def formatea_text(texto):
    n= texto.find('[')
    if  n == -1 :
        texto = texto.replace('\n\t','')
        texto = texto.replace('\n','')
    else:
        texto = texto[0:n]
                
    return texto

# indicar columnas a limpiar
columnas_a_limpiar = ['NOM_EQUIPO', 'CIUDAD', 'ENTRENADOR','CAPITAN','ESTADIO']

for col in columnas_a_limpiar:
    datos[col] = datos.apply(lambda fila: formatea_text(fila[col]), axis=1)

##==============================================================================
## convertir a entero el numero de AFORO de los estadios.
import re
def formatea_num(numero):
    if numero is None:
            return None
    numero = int(''.join(re.findall('[0-9]+', numero))) 
    
    return numero

datos.AFORO = datos.apply( lambda f: formatea_num(f.AFORO) , axis=1)

##==============================================================================
##IDENTIFICANDO LA COMUNIDAD < en este caso son pocos registros por ello se decisió hacerlo de esta manera>

datos['COMUNIDAD'] = datos.CIUDAD
def identifica_comunidad(texto1): 
    n = texto1.strip()
    if  n== 'Vitoria': texto1 = "pais_vasco"
    elif n=='Bilbao': texto1 = "pais_vasco"
    elif n=='San Sebastián': texto1 = "pais_vasco"
    elif n=='Éibar': texto1 = "pais_vasco"
    #------------------------------------
    elif n=='Madrid': texto1 = "com_madrid"
    elif n=='Getafe': texto1 = "com_madrid"
    elif n=='Leganés': texto1 = "com_madrid"
    #------------------------------------
    elif n=='Barcelona': texto1 = "cataluna"
    elif n=='Girona': texto1 = "cataluna"
    elif n=='Cornellá de Llobregat': texto1 = "cataluna"
    #------------------------------------
    elif n=='Sevilla': texto1 = "andalucia"
    #------------------------------------
    elif n=='Vigo': texto1 = "galicia"
    #------------------------------------
    elif n=='Huesca': texto1 = "aragon"
    #------------------------------------
    elif n=='Valencia': texto1 = "com_valenciana"
    elif n=='Villarreal': texto1 = "com_valenciana"
    #------------------------------------
    elif n=='Valladolid': texto1 = "castilla_y_leon" 
    else: 'NONE'
    return  texto1 

columnas_limpiar = ['COMUNIDAD']
for col in columnas_limpiar:
    datos[col] = datos.apply(lambda fila: identifica_comunidad(fila[col]), axis=1)


##=================================================================================
# Homogenizando los nombres de EQUIPOS

def formatea_equipo(texto):
    texto = texto.strip()
    texto = texto.replace('.','')
    texto = texto.lower()
    if texto.find('atlético') != -1:texto = "atletico_de_madrid" 
    elif texto.find('valencia') != -1:texto = "valencia_cf"  
    elif texto.find('barcelona') != -1:texto = "fc_barcelona"  
    elif texto.find('athletic') != -1:texto = "athletic_club"     
    elif texto.find('celta') != -1:texto = "celta_de_vigo"  
    elif texto.find('betis') != -1:texto = "real_betis" 
    elif texto.find('español') != -1:texto = "rcd_espanyol"        
    elif texto.find('leganés') != -1:texto = "leganes"   
    elif texto.find('getafe') != -1:texto = "getafe" 
    elif texto.find('girona') != -1:texto = "girona"   
    elif texto.find('alavés') != -1:texto = "alaves"        
    elif texto.find('huesca') != -1:texto = "huesca"   
    elif texto.find('eibar') != -1:texto = "sd_eibar"
    elif texto.find('levante') != -1:texto = "levante"   
    elif texto.find('valladolid') != -1:texto = "valladolid"  
    elif texto.find('rayo') != -1:texto = "rayo_vallecano"  
    elif texto.find('madrid') != -1:texto = "real_madrid"  
    elif texto.find('sevilla') != -1:texto = "sevilla_fc" 
    elif texto.find('villarreal') != -1:texto = "Villarreal"  
    else: texto
    return texto

datos['NOM_EQUIPO'] = datos.apply( lambda f: formatea_equipo(f.NOM_EQUIPO) , axis=1) 

##=================================================================================

datos.head() ## visualización de datos limpios
```

```python
## Dar formato numérico a los valores de mercado de equipo y de promedio por jugador (valore en euros)
##=====================================================================================================

datos1 = datos2.copy()   ## PUNTO DE CONTROL PARA EVITAR ERRORES Y PODER EJECUNAR N VECES

#---------------------------------------------------------------------------------------

def formatea_divisa(texto1):
    texto1 = texto1.strip()
    if texto1.find('Bill') == -1:
        texto1 = float('.'.join(re.findall('[0-9]+', texto1)))
    else:
        texto1 = float('.'.join(re.findall('[0-9]+', texto1)))*1000
    return texto1

# formatear columnas a in o float segun corresponda
columnas_num = ['VALOR_MER_EQ','VM_PROM_JUGADOR'] 

for col in columnas_num:
    datos1[col] = datos1.apply(lambda fila: formatea_divisa(fila[col]), axis=1)



## función para formatear los campos numéricos
##============================================
import re
def formatea_num(numero):
    
    if numero.find(',') == -1:
        numero = int(numero)
    else:
        numero = numero.replace(',','.')
        numero = float(numero)
    return numero

# formatear columnas a in o float segun corresponda
columnas_num = ['NUM_JUGADORES','EDAD_PROM', 'NUM_EXTRANJEROS'] 

for col in columnas_num:
    datos1[col] = datos1.apply(lambda fila: formatea_num(fila[col]), axis=1)

    

## Formateando los nombres de equipos para homogenizar datos
##============================================================
def formatea_equipo(texto):
    texto = texto.strip()
    texto = texto.replace(' ','_')
    texto = texto.lower()
    if texto.find('atlético') != -1:texto = "atletico_de_madrid" 
    elif texto.find('valencia') != -1:texto = "valencia_fc"        
    elif texto.find('athletic') != -1:texto = "athletic_club"       
    elif texto.find('espanyol') != -1:texto = "rcd_espanyol"        
    elif texto.find('leganés') != -1:texto = "leganes"          
    elif texto.find('alavés') != -1:texto = "alaves"        
    elif texto.find('huesca') != -1:texto = "huesca"              
    elif texto.find('valladolid') != -1:texto = "valladolid"                 
    else: texto
    return texto

datos1['NOM_EQUIPO'] = datos1.apply( lambda f: formatea_equipo(f.NOM_EQUIPO) , axis=1) 
datos1['UNI_MONEDA'] = "millones"
datos1['DIVISA'] = 'euros'
datos1.head()
```
